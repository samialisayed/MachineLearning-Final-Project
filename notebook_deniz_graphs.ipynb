```python
def plot_commit():
    from matplotlib.ticker import FuncFormatter
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib
    x = np.arange(3)
    money = [35,26,21]

    fig, ax = plt.subplots()
    plt.bar(x, money)
    plt.xticks(x, ('Tova', 'Sami', 'Deniz'))
    plt.title("Commits")
    plt.rcParams["font.size"] = 15
    plt.show()
```

```python
def plot_github():
    from matplotlib.ticker import FuncFormatter
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib
    labels = ['Deniz', 'Sami', 'Tova']
    additions = [41930, 19379, 18053]
    deletions = [2849, 3629, 2539]

    x = np.arange(len(labels))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, additions, width, label='Additions')
    rects2 = ax.bar(x + width/2, deletions, width, label='Deletions')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_title('Additions and Deletions')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    

```

```python
plot_commit()
```

```python
plot_github()
```
